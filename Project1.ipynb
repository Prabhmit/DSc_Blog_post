 ### Q3. Have the top five popular programming languages changed from year to year since 2011?
 
The number of programming languages used by participants are of a relatively large number. A high number of responses include
multiple combination of specific programming languages in the responses. For a clearer analysis, a list of 41 languages has been created below. For the analysis part, there were eight programming languages that had registered in yearly top five since 2011 which have been visualised.

```python
#List for possible programming languages
possible_vals = ['C#','Java','JavaScript','CSS','PHP','Python','Ruby','SQL','C#','C++','C','Perl','Haskell',
                 'VB','VBA','MATLAB','.NET','ABAP','ANDROID','Bash','Cobol','Delphi','Flash','FORTRAN','HTML','Swift',
                 'PHP','R','Shell','SCALA','VB.NET', 'SAS','Kotlin','Julia','PowerShell','Cassandra','Visual Basic',
                 'TypeScript','Go','Dart','Objective-C']
```

```python
len(possible_vals)
```

```python
#Function to calculate proportion of number of times programming language appears in column and add a year column to dataframe
def total_count(df, col1, list_name, year):
    '''
    INPUT:
    df - the pandas dataframe to search
    col1 - the column name  to look through
    list_name - a list of strings to search for in each row of df[col]
    
    OUTPUT:
    new_df - a dataframe of each list_name item with the count of how often it shows up 
    '''
    new_df1 = defaultdict(int)
    new_df2 = defaultdict(int)
    
    new_df1 = df[col1].value_counts().reset_index()
    new_df1.rename(columns={'index':'languages',col1:'count'},inplace=True)
    
    
    for val in list_name:
        for idx in range(new_df1.shape[0]):
            if val in new_df1['languages'][idx]:
                new_df2[val] += int(new_df1['count'][idx])   
    new_df2 = pd.DataFrame(pd.Series(new_df2)).reset_index()
    new_df2.columns = ['languages', 'count']
    new_df2.sort_values('count', ascending=False, inplace=True)
    new_df2['ratio'] = new_df2['count']/new_df2['count'].sum()
    new_df2['year'] = year
    return new_df2
```

Reference: jjrunner/stackoverflow/HowToBreakIntoTheField.ipynb, accessed 9 May 2020, <https://github.com/jjrunner/stackoverflow/blob/master/HowToBreakIntoTheField.ipynb>

```python
#2011-2015 data has multiple columns for programming languages with varying column titles. A list will be created for each.
#2018-2020 have the same name for programming languages column
#Function to join values from multiple columns and then apply the above function to calculate proportions and add year column

def combine_total_count(df,list_name,year):
    '''
    INPUT:
    df - the pandas dataframe to search
    col1 - the column name  to join and look through
    list_name - a list of strings to search for in each row of df[col]
    
    OUTPUT:
    new_df - a dataframe of each list_name item with its count  
    '''
    df['languages'] = df[list_name].apply(lambda x: ','.join(x[x.notnull()]), axis=1)
    return total_count(df, 'languages', possible_vals, year)   
```
